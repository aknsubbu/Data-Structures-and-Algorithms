# LCA in BST

```python
from random import randint
from time import perf_counter
import matplotlib.pyplot as plt
import tracemalloc


class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def insert(root, data):
    if root is None:
        return Node(data)
    elif data <= root.data:
        root.left = insert(root.left, data)
    else:
        root.right = insert(root.right, data)
    return root


def generate_bst(n):
    root = None
    for _ in range(n):
        data = randint(1, NUMBER - 1)
        root = insert(root, data)
    return root


def lowestCommonAncestor(root, p, q):
    def findCommon(node, p, q):
        if node is None:
            return None
        if p > node.data and q > node.data:
            return findCommon(node.right, p, q)
        elif p < node.data and q < node.data:
            return findCommon(node.left, p, q)
        else:
            return node

    return findCommon(root, p, q)


def analyze_function(func, bst, p, q):
    start_time = perf_counter()
    tracemalloc.start()
    func(bst, p, q)
    end_time = perf_counter()
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    time_taken = end_time - start_time
    memory_peak = peak / 10**6  # Convert to megabytes
    return time_taken, memory_peak


def plot_results(sizes, times, memories, title="Analysis"):
    plt.plot(sizes, times, label="Time")
    plt.plot(sizes, memories, label="Memory")
    plt.xlabel("BST Size")
    plt.ylabel("Time (seconds) / Memory Usage (MB)")
    plt.title(title)
    plt.legend()
    plt.show()


NUMBER = 1001
# Generate random BSTs
sizes = range(1, NUMBER)
bsts = [generate_bst(size) for size in sizes]

i = 1
# Time and memory analysis
results = [analyze_function(lowestCommonAncestor, bsts[size-1], randint(1, i), randint(1, i)) for size in sizes]

# Unpack the results into separate lists for plotting
times, memories = zip(*results)

# Plot the results
plot_results(sizes, times, memories)
```
